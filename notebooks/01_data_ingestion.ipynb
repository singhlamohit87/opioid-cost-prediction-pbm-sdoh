```python
import pandas as pd
import os
import glob

# Paths
geo_dir = "../data/cms/OpioidPrescribingRates_Geography/2023/"
provider_dir = "../data/cms/OpioidPrescribers_Provider/"
drug_geo_dir = "../data/cms/OpioidPrescribers_Geography_Drug/"
sdoh_path = "../data/2024_county_health_SDOH_data.xlsx"

years = list(range(2013, 2024))  # 2013–2023

# -------------------------------
# 1. Opioid Prescribing Rates (Geography) — only 2023
# -------------------------------
geo_files = glob.glob(os.path.join(geo_dir, "*.csv"))
if len(geo_files) == 0:
    raise FileNotFoundError("No 2023 GEO prescribing file found in folder.")
df_geo = pd.read_csv(geo_files[0])
df_geo["year"] = 2023
print("df_geo shape:", df_geo.shape)

# -------------------------------
# 2. Prescribers by Provider (multi-year)
# -------------------------------
presc_dfs = []
for y in years:
    files = glob.glob(os.path.join(provider_dir, str(y), "*.csv"))
    if not files:
        print(f"[MISSING] No provider file for {y}")
        continue
    f = files[0]
    tmp = pd.read_csv(f, encoding="latin1", low_memory=False)
    tmp["year"] = y
    presc_dfs.append(tmp)
    print(f"[LOADED] Provider {y} → {tmp.shape}")
df_prescribers = pd.concat(presc_dfs, ignore_index=True)
print("df_prescribers combined shape:", df_prescribers.shape)

# -------------------------------
# 3. Prescribers by Geography & Drug (multi-year)
# -------------------------------
drug_geo_dfs = []
for y in years:
    files = glob.glob(os.path.join(drug_geo_dir, str(y), "*.csv"))
    if not files:
        print(f"[MISSING] No drug_geo file for {y}")
        continue
    f = files[0]
    tmp = pd.read_csv(f, encoding="latin1", low_memory=False)
    tmp["year"] = y
    drug_geo_dfs.append(tmp)
    print(f"[LOADED] Drug+Geo {y} → {tmp.shape}")
df_drug_geo = pd.concat(drug_geo_dfs, ignore_index=True)
print("df_drug_geo combined shape:", df_drug_geo.shape)

# -------------------------------
# 4. SDOH snapshot (2024)
# -------------------------------
df_sdh = pd.read_excel(sdoh_path, sheet_name="Select Measure Data", skiprows=1)
print("df_sdh shape:", df_sdh.shape)

# -------------------------------
# 5. Save processed
# -------------------------------
os.makedirs("../data/processed", exist_ok=True)
df_geo.to_csv("../data/processed/geo_2023.csv", index=False)
df_prescribers.to_csv("../data/processed/prescribers_2013_2023.csv", index=False)
df_drug_geo.to_csv("../data/processed/drug_geo_2013_2023.csv", index=False)
df_sdh.to_excel("../data/processed/sdoh_2024.xlsx", index=False)

print("All files ingested and saved.")
```
